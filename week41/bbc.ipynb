1. Scrape https://www.bbc.com/ for "div p h3 a" tags
2. Sæt tags ind i en Dataframe - bestem selv hvordan den skal se ud.
3. Sorter og plot dem fra størst til mindst forekomster.


```python
import bs4
import requests
import pandas as pd

r = requests.get('https://www.bbc.com/')
r.raise_for_status()
soup = bs4.BeautifulSoup(r.text, 'html.parser')

div_tags = len(soup.findAll('div'))
p_tags = len(soup.findAll('p'))
h3_tags = len(soup.findAll('h3'))
a_tags = len(soup.findAll('a'))

tags_dict = {'div': div_tags, 'p': p_tags, 'h3': h3_tags, 'a':a_tags}

tags = pd.DataFrame(tags_dict, index=[0])

sorted = tags.sort_values(0, axis=1, ascending=False)
display(sorted)


sorted.plot(kind="bar")
```
